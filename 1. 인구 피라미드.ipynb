# 1. 인구 피라미드

## 남자 데이터

```python
import pandas as pd
file_name = '202303_202303_연령별인구현황_월간.xlsx'
df_m = pd.read_excel(file_name, skiprows=3, index_col = '행정기관', usecols='B,E:X') #index_col='행정기관'
df_m
```

```python
df_m.iloc[0] = df_m.iloc[0].str.replace(',','').astype(int)
```

```python
df_m.iloc[0]
```

# 여자 데이터

```python
df_w = pd.read_excel(file_name, index_col='행정기관',skiprows=3, usecols='B,AB:AU')
df_w.head(3)
```

```python
df_w.columns
```

```python
df_w.columns = df_m.columns # 컬럼명 통일
df_w.columns
```

```python
df_w.iloc[0]=df_w.iloc[0].str.replace(',','').astype(int)
```

```python
df_m.iloc[0]
```

## 데이터 시각화

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
# matplotlib.rcParams['font.family'] = "AppleGothic" Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용 시, 마이너스 글자가 깨지는 현상을 해결
```

```python
plt.figure(figsize=(9 ,6))
plt.barh(df_m.columns,-df_m.iloc[0] // 1000) #astype(int) 형변형
plt.barh(df_w.columns, df_w.iloc[0] // 1000)
plt.title('2023 대한민국 인구 피라미드')
plt.savefig('2023_인구피라미드')
plt.show()
```
