```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt 
```

```python
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 15
```

```python
data1 = pd.read_csv('./data/data1.csv')
data1
```

```python
m1t1 = data1.iloc[0:128,1:2]
m1t2 = data1.iloc[0:128,2:3]
m1t3 = data1.iloc[0:128,3:4]

m1t3
```

```python
plt.plot(m1t1, label='train/box_loss')
plt.plot(m1t2, label='train/obj_loss')
plt.plot(m1t3, label='train/cls_loss')

plt.legend(fontsize=9)
```

```python
m1v1 = data1.iloc[0:128,8:9]
m1v2 = data1.iloc[0:128, 9:10]
m1v3 = data1.iloc[0:128, 10:11]
```

```python
plt.plot(m1v1, label='val/box_loss')
plt.plot(m1v2, label='val/obj_loss')
plt.plot(m1v3, label='val/cls_loss')

plt.legend(fontsize=9)
```

```python
data2 = pd.read_csv('./data/data2.csv')
data2
```

```python
m2t1 = data2.iloc[0:128,1:2]
m2t2 = data2.iloc[0:128,2:3]
m2t3 = data2.iloc[0:128,3:4]
m2v1 = data2.iloc[0:128,8:9]
m2v2 = data2.iloc[0:128, 9:10]
m2v3 = data2.iloc[0:128, 10:11]
```

```python
plt.plot(m1t1,color='cadetblue',label='data1_train/box_loss')
plt.plot(m2t1,color='powderblue',label='data2_train/box_loss')
plt.plot(m1t2,color='darkseagreen',label='data1_train/obj_loss')
plt.plot(m2t2,color='lightgreen',label='data2_train/obj_loss')
plt.plot(m1t3,color='tomato',label='data1_train/cls_loss')
plt.plot(m2t3,color='lightcoral',label='data2_train/cls_loss')
plt.legend()
plt.savefig('sex.png',dpi=100)
```

```python
plt.plot(m1t1,color='teal',label='data1')
plt.plot(m2t1,color='lightcoral',label='data2')
plt.legend()
plt.title("box_loss")
plt.savefig('box_loss.png',dpi=100)
```

```python
plt.plot(m1t2,color='teal',label='data1')
plt.plot(m2t2,color='lightcoral',label='data2')
plt.legend()
plt.title("obj_loss")
plt.savefig('obj_loss.png',dpi=100)
```

```python
plt.plot(m1t3,color='teal',label='data1')
plt.plot(m2t3,color='lightcoral',label='data2')
plt.legend()
plt.title("cls_loss")
plt.savefig('clss_loss',dpi=100)
```
