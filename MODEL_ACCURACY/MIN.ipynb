```python
# 주어진 데이터
box = ['data1', 'data2']
box_mins = [0.028473, 0.025609]

obj = ['data1', 'data2']
obj_mins = [0.016647, 0.01648]

cls = ['data1', 'data2']
cls_mins = [0.000931, 0.000771]

# 데이터를 DataFrame으로 변환
data = pd.DataFrame({
    'Type': ['box'] * len(box) + ['obj'] * len(obj) + ['cls'] * len(cls),
    'Category': box + obj + cls,
    'Values': box_mins + obj_mins + cls_mins,
})

# seaborn의 스타일 설정
sns.set_theme(style="whitegrid")

# catplot 그리기
g = sns.catplot(x="Type", y="Values", hue="Category", kind="bar", data=data, height=6, aspect=1.2,
                errcolor=".2", edgecolor=".2", palette="dark", alpha=.6)
g.despine(left=True)
g.set_axis_labels("", "최소값")
plt.ylim(0, 0.03)  # y축 범위 조절
plt.title("Minimum Values by Type and Category")

# 각 막대 위에 값 표시
ax = g.facet_axis(0, 0)
for p in ax.patches:
    ax.text(p.get_x() + p.get_width() / 2., p.get_height() + 0.0003, f'{p.get_height():.6f}', ha='center', va='bottom')


plt.show()
```
